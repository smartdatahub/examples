```python
import pandas as pd
```

```python
# Copied & pasted from
# https://www.smartdatahub.io/dataset/fi_museovirasto_arliga_tabeller_2022_xlsx --> Integrate --> Python pandas
# For column descriptions, see https://www.smartdatahub.io/dataset/fi_museovirasto_arliga_tabeller_2022_xlsx --> Details
from sdhpy.pandas import SdhPandas
sdh_catalog = SdhPandas(store_apikey="xxxxxxxx-xxxx-xxxx-xxxx-xxxxxxxxxxxx")
sdh_dataset = sdh_catalog.fi_museovirasto.arliga_tabeller_2022_xlsx__kaikkien_vastaajien_kysymykset
df = sdh_dataset.data
df.head()
```

```python
subsidy_efficiency = df.rename(columns = {"museets_namn": "Museum", \
                                          "alla_besok_sammanlagt_det_totala_besokarantalet": "Total Visitors", \
                                          "statsunderstod": "Government Subsidy"}) \
   [["Museum", "Total Visitors", "Government Subsidy"]]
subsidy_efficiency["Visitor Subsidy Efficiency Ratio"] = subsidy_efficiency["Total Visitors"].div(subsidy_efficiency["Government Subsidy"])
pd.set_option('use_inf_as_na', True)
subsidy_efficiency.dropna(subset=["Visitor Subsidy Efficiency Ratio"]).sort_values(by=["Visitor Subsidy Efficiency Ratio"], ascending=False)
```
